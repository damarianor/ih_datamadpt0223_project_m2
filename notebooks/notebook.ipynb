```python
dataset = pd.read_csv('../data/us_mass_shootings_2.csv')
dataset
```

```python
df_normalize = dataset.copy()
df_normalize
```

```python
df_city = dataset['City'] 
df_city = df_city.drop_duplicates()
df_city
```

```python
new_df_city = df_city.drop_duplicates().reset_index(drop=True)
new_df_city = pd.DataFrame(new_df_city)
new_df_city
```

```python
print(new_df_city.columns)
```

```python
new_df_city = new_df_city.reset_index()
new_df_city = new_df_city.rename(columns={'index': 'id_city'})
new_df_city
```

```python
print(new_df_city.columns)
```

```python
df_gender = dataset['Gender'] 
df_gender = df_gender.drop_duplicates()
df_gender
```

```python
new_df_gender = df_gender.drop_duplicates().reset_index(drop=True)
new_df_gender = pd.DataFrame(new_df_gender)
new_df_gender
```

```python
new_df_gender = new_df_gender.reset_index()
new_df_gender = new_df_gender.rename(columns={'index': 'id_gender'})
new_df_gender
```

```python
df_location = dataset['Location'] 
df_location = df_location.drop_duplicates()
df_location
```

```python
new_df_location = df_location.drop_duplicates().reset_index(drop=True)
new_df_location = pd.DataFrame(new_df_location)
new_df_location
```

```python
new_df_location = new_df_location.reset_index()
new_df_location = new_df_location.rename(columns={'index': 'id_location'})
new_df_location
```

```python
df_race = dataset['Race'] 
df_race = df_race.drop_duplicates()
df_race
```

```python
new_df_race = df_race.drop_duplicates().reset_index(drop=True)
new_df_race = pd.DataFrame(new_df_race)
new_df_race
```

```python
new_df_race = new_df_race.reset_index()
new_df_race = new_df_race.rename(columns={'index': 'id_race'})
new_df_race
```

```python
df_merge = pd.merge(df_normalize, new_df_city, how='inner', on = 'City')
df_merge
```

```python
df_merge_2 = pd.merge(df_merge, new_df_gender, how='inner', on = 'Gender')
df_merge_2
```

```python
df_merge_3 = pd.merge(df_merge_2, new_df_location, how='inner', on = 'Location')
df_merge_3
```

```python
df_merge_4 = pd.merge(df_merge_3, new_df_race, how='inner', on = 'Race')
df_merge_4
```

```python
df_merge_4
```

```python
df_merge_4.to_csv('dataframe_normalize.csv', index=False)
```
